Read all tables from volume and write into the bronze layer

```python
filepath_dict = {
  "workspace.bronze.olist_customers" :"/Volumes/workspace/olist/raw_data/olist_customers_dataset.csv",
  "workspace.bronze.olist_geolocation" : "/Volumes/workspace/olist/raw_data/olist_geolocation_dataset.csv",
  "workspace.bronze.olist_order_items" : "/Volumes/workspace/olist/raw_data/olist_order_items_dataset.csv",
  "workspace.bronze.olist_order_payments" :"/Volumes/workspace/olist/raw_data/olist_order_payments_dataset.csv" ,
  "workspace.bronze.olist_order_reviews" :"/Volumes/workspace/olist/raw_data/olist_order_reviews_dataset.csv",
  "workspace.bronze.olist_orders" : "/Volumes/workspace/olist/raw_data/olist_orders_dataset.csv",
  "workspace.bronze.olist_products" : "/Volumes/workspace/olist/raw_data/olist_products_dataset.csv", 
  "workspace.bronze.olist_sellers" :"/Volumes/workspace/olist/raw_data/olist_sellers_dataset.csv", 
  "workspace.bronze.olist_product_category_name_translation":"/Volumes/workspace/olist/raw_data/product_category_name_translation.csv"
}
```

```python
for key, values in filepath_dict.items():
    df = spark.read.option("header","true").option("inferSchema","true").csv(values)
    # Use the full table path from the dictionary key
    df.write.mode("overwrite").saveAsTable(f"{key}")
```
